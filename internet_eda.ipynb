**4. RELACIÓN ENTRE FEATURES**

**4.1.** Respecto al dataframe df1:

Diagrama de dispersión:

```python
df1.plot(kind = 'scatter',
        x = '1 mbps',
        y = '2 mbps',
        title = '1mbps vs 2mbps')
plt.show()
```

Diagrama de dispersión con hue:

```python
# Antes convertiremos los dtype de 'Link Indec' a float:
lista_flotantes = []
for d in df1['Link Indec']:
    if d == 'Sin Datos':
        d = np.nan
        lista_flotantes.append(d)
    else:
        d = float(d)
        d = lista_flotantes.append(d)
df1['Link Indec'] = lista_flotantes
```

```python
ax = sns.scatterplot(x = '1 mbps',
                    y = '2 mbps',
                    hue = 'Link Indec',
                    data = df1)
ax.set_title('1 mbps vs 2mbps')
plt.show()
```

Gráfico por pares con hue:

```python
sns.pairplot(df1,
             vars = ['1 mbps', '2 mbps',
                     '3 mbps', '4 mbps', '5 mbps'],
            hue = 'Link Indec')
plt.show()
```

Matriz de correlación:

```python
df1_corr = df1[['1 mbps', '2 mbps', '3 mbps',
                '4 mbps', '5 mbps']].dropna().corr()
df1_corr
```

Mapa de calor de correlaciones:

```python
sns.heatmap(df1_corr, annot = True)
```

**5. PREGUNTA RESPONDIDA SOBRE LOS DATOS**

¿Cuáles son las localidades que tienen mayor velocidad de internet?

```python
s = df1.sort_values(by = '100 mbps', ascending = False)
e = s.head(10)

# Establecemos el tamaño de la gráfica:
plt.figure(figsize = (10,6))

# Creamos la gráfica:
plt.bar(e['Localidad'], e['100 mbps'])

# Establecemos el nombre de las etiquetas:
plt.xlabel('Localidad')
plt.ylabel('Cantidad de familias con 100 mbps de velocidad')
plt.title('Top 10')
#Establecemos la visualización de etiquetas del eje x en formato vertical:
plt.xticks(rotation = 90)

plt.show()
```
