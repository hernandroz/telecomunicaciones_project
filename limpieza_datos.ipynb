**6. ASIGNACIÓN DE DTYPE CORRECTO**

Respecto al dataframe df1:

```python
df1.head(3)
```

```python
print('Lista ordenada de los dtypes de las columnas:')
print(list(df1.dtypes))
```

```python
# Cambiaremos a str el dtype de "Provincia", "Partido", "Localidad" y "Link Indec":
df1['Provincia'] = df1['Provincia'].astype(str)
df1['Partido'] = df1['Partido'].astype(str)
df1['Localidad'] = df1['Localidad'].astype(str)
df1['Link Indec'] = df1['Link Indec'].astype(str)
```

```python
# Para las demás conversiones, primero imprimiremos los nombres de
# las columnas para ayudarnos con la siguiente celda:
df1.columns
```

```python
# Convertiremos todos los valores de las columnas restantes a tipo float. 
# Teniendo en cuenta que hay valores del tipo "28,837", se reemplazará
# el caracter "," por "." para hacer efectivo el cambio de dtype:

for c in ['0,256 mbps', '0,375 mbps', '0,5 mbps', '0,512 mbps', '0,625 mbps',
       '0,75 mbps', '1 mbps', '1,25 mbps', '1,5 mbps', '2 mbps', '2,2 mbps',
       '2,5 mbps', '3 mbps', '3,3 mbps', '3,5 mbps', '4 mbps', '4,5 mbps',
       '5 mbps', '6 mbps', '6,4 mbps', '7 mbps', '7,5 mbps', '8 mbps',
       '9 mbps', '10 mbps', '11 mbps', '12 mbps', '13 mbps', '14 mbps',
       '15 mbps', '16 mbps', '17 mbps', '18 mbps', '19 mbps', '20 mbps',
       '21 mbps', '22 mbps', '23 mbps', '24 mbps', '25 mbps', '25,1 mbps',
       '25,11 mbps', '25,5 mbps', '26 mbps', '30 mbps', '31 mbps', '32 mbps',
       '34 mbps', '35 mbps', '36 mbps', '38 mbps', '39 mbps', '40 mbps',
       '41 mbps', '45 mbps', '46 mbps', '49 mbps', '50 mbps', '55 mbps',
       '58 mbps', '59 mbps', '60 mbps', '61 mbps', '62 mbps', '64 mbps',
       '65 mbps', '66 mbps', '70 mbps', '75 mbps', '78 mbps', '80 mbps',
       '82 mbps', '83 mbps', '85 mbps', '90 mbps', '92 mbps', '95 mbps',
       '100 mbps']:
    lista_flotantes = []
    for d in df1[c]:
        try:
            d = float(d)
            lista_flotantes.append(d)
        except ValueError:
            d = d.replace(",", ".")
            d = float(d)
            lista_flotantes.append(d)
    df1[c] = lista_flotantes
```

```python
# Comprobamos que las conversiones se hayan realizado con éxito:
print('Lista ordenada de los dtypes de las columnas:')
print(list(df1.dtypes))
```

```python
# Imprimimos los dtypes de las columnas:
df2.dtypes
```

```python
# Hacemos la conversión de dtype de la columna 'Latitud', considerando que existen valores  
# del tipo '-38,6472605094596' a los que previamente reemplazaremos el caracter ',' por '.'

lista_flotantes = []
for d in df2['Latitud']:
    try:
        d = float(d)
        lista_flotantes.append(d)
    except ValueError:
        try:
            d = d.replace(',', '.')
            d = float(d)
            lista_flotantes.append(d)
        except ValueError:
            d = d.replace('..', '.')
            d = float(d)
            lista_flotantes.append(d)
df2['Latitud'] = lista_flotantes
```

```python
# Ahora hacemos la conversión de dtype de la columna 'Longitud', considerando que
# también en ella existen valores del tipo '-62,3200223419024' a los cuales 
# previamente reemplazaremos el caracter ',' por '.'

lista_flotantes = []
for d in df2['Longitud']:
    try:
        d = float(d)
        lista_flotantes.append(d)
    except ValueError:
        try:
            d = d.replace(',', '.')
            d = float(d)
            lista_flotantes.append(d)
        except ValueError:
            d = d.replace('..', '.')
            d = float(d)
            lista_flotantes.append(d)
df2['Longitud'] = lista_flotantes
```

```python
# Finalmente, comprobamos que las conversiones se hayan realizado con éxito:
df2.dtypes
```
